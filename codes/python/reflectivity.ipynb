```python
import numpy as np
from ROOT import TFile, TH1D, TCanvas, kRed, kBlue, kGreen, gPad, TLegend
import IPython.display as disp
```

```python
myfile2=TFile.Open("/home/tushar/Normal-work/data/reflectivity_v1/per20.root","READ")
myfile3=TFile.Open("/home/tushar/Normal-work/data/reflectivity_v4/per20.root","READ")

mytree2=myfile2["Scoring"]
mytree3=myfile3["Scoring"]

h2p1 =  TH1D("h2p1","Detector v1 PMT1 response",60,0,60);
h3p1 =  TH1D("h3p1","Detector v4 PMT1 response",60,0,60);

h2p2 =  TH1D("h2p2","Detector v1 PMT2 response",60,0,60);
h3p2 =  TH1D("h3p2","Detector v4 PMT2 response",60,0,60);

h2p3 =  TH1D("h2p3","Detector v1 PMT3 response",60,0,60);
h3p3 =  TH1D("h3p3","Detector v4 PMT3 response",60,0,60);


for entry in mytree2:
    h2p1.Fill(entry.c4PMT0SC);
    h2p2.Fill(entry.c5PMT1SC);
    h2p3.Fill(entry.c6PMT2SC);
for entry in mytree3:
    h3p1.Fill(entry.c4PMT0SC);
    h3p2.Fill(entry.c5PMT1SC);
    h3p3.Fill(entry.c6PMT2SC);
    
t1= TCanvas("t1","Canvas",800,600)
legend1 = TLegend();
legend1.AddEntry(h2p1,"Detector v1");
legend1.AddEntry(h3p1,"Detector v4");
#h2p1.SetFillStyle(3144);
#h3p1.SetFillStyle(3305);
h2p1.SetFillColorAlpha(kRed,0.65);
h3p1.SetFillColorAlpha(kBlue,0.65);
h2p1.Draw();
h3p1.Draw("SAME");
legend1.Draw();
h2p1.SetTitle("PMT1 response for 20% reflectivity");
h2p1.SetXTitle("No of photons recorded");
t1.Draw();
disp.display(t1);

t2= TCanvas("t2","Canvas",800,600)
legend2 = TLegend();
legend2.AddEntry(h2p2,"Detector v1");
legend2.AddEntry(h3p2,"Detector v4");
#h2p1.SetFillStyle(3144);
#h3p1.SetFillStyle(3305);
h2p2.SetFillColorAlpha(kRed,0.65);
h3p2.SetFillColorAlpha(kBlue,0.65);
h2p2.Draw();
h3p2.Draw("SAME");
legend2.Draw();
h2p2.SetTitle("PMT2 response for 20% reflectivity");
h2p2.SetXTitle("No of photons recorded");
t2.Draw();
disp.display(t2);

t3= TCanvas("t3","Canvas",800,600)
legend3 = TLegend();
legend3.AddEntry(h2p3,"Detector v1");
legend3.AddEntry(h3p3,"Detector v4");
#h2p1.SetFillStyle(3144);
#h3p1.SetFillStyle(3305);
h2p3.SetFillColorAlpha(kRed,0.65);
h3p3.SetFillColorAlpha(kBlue,0.65);
h2p3.Draw();
h3p3.Draw("SAME");
legend3.Draw();
h2p3.SetTitle("PMT2 response for 20% reflectivity");
h2p3.SetXTitle("No of photons recorded");
t3.Draw();
disp.display(t3);
```
